```python
labels = [{'name':'ThumbsUp', 'id':1}, {'name':'ThankYou', 'id':2}, {'name':'Boo', 'id':3}, {'name':'LiveLong', 'id':4}]

with open(files['LABELMAP'], 'w') as f:
    for label in labels:
        f.write('item { \n')
        f.write('\tname:\'{}\'\n'.format(label['name']))
        f.write('\tid:{}\n'.format(label['id']))
        f.write('}\n')
```
